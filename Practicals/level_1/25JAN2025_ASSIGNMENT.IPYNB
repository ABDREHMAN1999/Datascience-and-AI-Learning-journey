```python
def mark_attendence():
    students = set()
    while True:
        student = input("Enter student name: ")
        students.add(student)
        if "STOP" in students:
            students.remove("STOP")
            break
    return students , len(students)

(a , b) = mark_attendence()
print(f"the list of students present today is {a}")
print(f"the total number of students present today is {b}")
```
