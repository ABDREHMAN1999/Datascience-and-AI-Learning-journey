```python
def our_path_maker(os ="windows", *args):
    path = ""
    for item in args:
        if os =="windows":
            path = path + item + "/"
        elif os == "Mac":
            path = path + item + "\\\\"
    return path 

path = our_path_maker("Mac", "abdul", "datascience", "new")
print(path)
```
