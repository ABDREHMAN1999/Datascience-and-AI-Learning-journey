```python
class ultimatedatascience:
    def __init__(self, batch):
        self.batch = batch
        self.students = []
        
    def AddStudent(self, name, email):
        self.name = name 
        self.email = email 
        details = {"name":name, "email": email}
        self.students.append(details)
        print(f"{self.name} is added to {self.batch}")
        
    def print_student_name(self):
        print(f"Student list for {self.batch} :")
        for student in self.students:
            print(f"- {student["name"]}")
            
            
s1 = ultimatedatascience("batch1")
s1.AddStudent("Anup", "Anup@gmail.com")
s1.print_student_name()
s1.AddStudent("Abdulrehman", "rehmaniabdul.shaikh@gmail.com")
s1.print_student_name()

s2 = ultimatedatascience("batch2")
s2.AddStudent("Ganesh", "Ganesh@gmail.com")
s2.AddStudent("Kaushik", "Kaushik@gmail.com")
s2.print_student_name()
        
            
        
        
        
    
    
        
        
        
        
        
```
